### VENN DIAGRAMS

```python
import matplotlib.pyplot as plt
import numpy as np

params={'xtickslabels':[],'ytickslabels':[]}
def circle(center,r):
    theta = np.linspace(0,2*np.pi,2000)
    x=r*np.cos(theta)+center[0]
    y=r*np.sin(theta)+center[1]
    plt.plot(x,y,c='k')
    return x,y

def fill_between(x1,y1,x2,y2,op):
    if op == 'interception':
        x=x1[x1 >= np.hstack([x2[len(x2)//2:],x2[:len(x2)//2]])]
        y=y1[x1 >= np.hstack([x2[len(x2)//2:],x2[:len(x2)//2]])]
        plt.fill_between(x,y,facecolor='b')
        x=x2[x2 <= np.hstack([x1[len(x1)//2:],x1[:len(x1)//2]])]
        y=y2[x2 <= np.hstack([x1[len(x1)//2:],x1[:len(x1)//2]])]
        plt.fill_between(x,y,facecolor='b')
    if op == 'union':
        plt.fill_between(x1,y1,facecolor='b')
        plt.fill_between(x2,y2,facecolor='b')
    if op == 'complement':
        plt.fill_between([-0.5,9],[5.5,5.5],facecolor='b')
        plt.fill_between([-0.5,9],[-1.5,-1.5],facecolor='b')
        plt.fill_between(x1,y1,facecolor='w')
        plt.fill_between(x2,y2,facecolor='w') 
    if op == 'difference':
        plt.fill_between(x1,y1,facecolor='b')
        plt.fill_between(x2,y2,facecolor='w') 
plt.style.use('seaborn-white')
fig = plt.subplots(figsize = (16,12),dpi=80)

# intersection
plt.subplot(3,3,1,ylim=[-1.5,5.5],xlim=[-0.5,9])
plt.axis('off')
this='B\\cap A           '
plt.title('$%s$'%this,fontsize=16)
x1, y1 = circle((2,2),2)
x2, y2 = circle((5,2),2)
fill_between(x1,y1,x2,y2,'interception')

# Union
plt.subplot(3,3,2,ylim=[-1.5,5.5],xlim=[-0.5,9])
plt.axis('off')
this='B\\cup A           '
plt.title('$%s$'%this,fontsize=16)
x1, y1 = circle((2,2),2)
x2, y2 = circle((5,2),2)
fill_between(x1,y1,x2,y2,'union')

# Union Complement
plt.subplot(3,3,3,ylim=[-1.5,5.5],xlim=[-0.5,9])
plt.xticks([])
plt.yticks([])
this='(B\\cup A)^c           '
plt.title('$%s$'%this,fontsize=16)
x1, y1 = circle((2,2),2)
x2, y2 = circle((5,2),2)
fill_between(x1,y1,x2,y2,'complement')

# difference
plt.subplot(3,3,4,ylim=[-1.5,5.5],xlim=[-0.5,9])
plt.axis('off')
plt.title('A-B             ',fontsize=16)
x1, y1 = circle((2,2),2)
x2, y2 = circle((5,2),2)
fill_between(x1,y1,x2,y2,'difference')
plt.subplot(3,3,5,ylim=[-1.5,5.5],xlim=[-0.5,9])
plt.axis('off')
plt.title('B-A             ',fontsize=16)
x1, y1 = circle((2,2),2)
x2, y2 = circle((5,2),2)
fill_between(x2,y2,x1,y1,'difference')
```
